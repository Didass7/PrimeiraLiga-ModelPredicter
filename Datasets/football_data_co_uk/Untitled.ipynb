```python
import pandas as pd
import glob
import os
```

```python
caminho= "/Diogo/Ambiente de Trabalho/PROJETO/Datasets/football_data_co_uk/*.csv"
ficheiros= glob.glob(caminho)
lista_dfs = []

for filename in ficheiros:
    df = pd.read_csv(filename)
    lista_dfs.append(df)

df_jogos = pd.concat(lista_dfs, axis=0, ignore_index=True)
```

```python
df_jogos['Date'] = pd.to_datetime(df_jogos['Date'], dayfirst=True)
df_jogos = df_jogos.sort_values('Date')
print(df_jogos.head())
```
